```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
#IMPORTING THE DATA
car_data = pd.read_csv("car.data")
```

```python
car_data.head(10)
```

### Converting categorical variables into label codes

```python
from sklearn.preprocessing import LabelEncoder
lbc = LabelEncoder()
car_data["buying"] = lbc.fit_transform(car_data["buying"])
car_data["maint"] = lbc.fit_transform(car_data["maint"])
car_data["door"] = lbc.fit_transform(car_data["door"])
car_data["persons"] = lbc.fit_transform(car_data["persons"])
car_data["lug_boot"] = lbc.fit_transform(car_data["lug_boot"])
car_data["safety"] = lbc.fit_transform(car_data["safety"])
car_data["class"] = lbc.fit_transform(car_data["class"])
```

```python
car_data.head()
```

```python
car_data['class'].unique()
```

### Defining X and y

```python
X = car_data.drop(['class'],axis =1).values
y = car_data['class'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state =0)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
y1 = lbc.fit_transform(car_data["class"])
rmse=[]
i=0
for k in range(20):
    k=k+1
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X,y1)
#y_pred_knn=knn.predict(X)
    rmse.append(sqrt(mean_squared_error(y1,knn.predict(X))))
    print('K value: ',k ,'Rmse :',rmse[k-1])

```

```python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
```

```python
acc = knn.score(X_test,y_test)
print(acc)
```
